### **Ejercicio 1**

Crea un array bidimensional de numpy con los siguientes valores:
\begin{bmatrix}
10 & 15 & 20 & 25 \\
30 & 35 & 40 & 45
\end{bmatrix}
y, a continuación:

- **a)** Muestra por pantalla:
  - El número de dimensiones del array.  
  - La forma (*shape*) del array.  
  - El número total de elementos.  
  - El valor del elemento situado en la **primera fila y tercera columna**.  
  - Los **dos últimos elementos** de la **primera fila**.  
  **(0.25 puntos)**

- **b)** Recorre todos los elementos del array y muestra únicamente aquellos que sean **múltiplos de 5 y mayores que 20**.  
  Indica también cuántos elementos cumplen esa condición.  
  **(0.25 puntos)**

**(Total: 0.5 puntos)**  
$\color{green}{\text{NM}}$

```python
#Importamos libs y creamos array
import numpy as np

arraybi = np.array([[10, 15, 20, 25],
                  [30, 35, 40, 45]])


```

```python
# Apartado A

#Imprimimos el número de dimensiones del array

print(f"Número de dimensiones del array: {arraybi.ndim}")

#Imprimimos la forma (shape) del array

print(f"Forma del array: {arraybi.shape}")

#Número total de elementos

print(f"Número total de elementos: {arraybi.size}")

#Imprimimos el elemento de la primera fila y tercera columna.

print(f"El elemento en la primera fila y tercera columna es = {arraybi[0, 2]}")

#Por último, los dos últimos elementos de la primera fila, utilizamos (-2:) en
#columna para seleccionar todos los elementos a partir del penúltimo, es decir,
#los dos últimos.
#Para imprimirlos, accedemos a cada elemento por separado para que sea más visual.

dos_ultimos = arraybi[0, -2:]
print(f"Los dos últimos elementos de la primera fila son: {dos_ultimos[0]} y {dos_ultimos[1]}")

#Como hemos hecho un slicing y asignado a la variable dos_ultimos el subarray con
#los dos últimos elementos, es decir dimensión 1, podemos acceder a cada elemento
#simplemente con un sólo número.

#Alternativamente:
print(f"Los dos últimos elementos de la primera fila son: {arraybi[0, -2:]}")
```

```python
# Apartado B
#Primero creamos la variable contador que nos va a servir para la parte final,
#indicar cuantos elementos cumplen la condición.
contador = 0
#Con dos for vamos recorriendo cada fila, y dentro de cada fila, cada columna.
#utilizamos .shape que devuelve una tupla cuyos elementos son número de filas y
#número de columnas, por eso accedemos con posición 0 y 1 a cada.
for i in range(arraybi.shape[0]):
    for j in range(arraybi.shape[1]):
#Asignamos a la variable valor el elemento i,j y verificamos si cumple la condición
        valor = arraybi[i, j]
        if valor % 5 == 0 and valor > 20:
            print(valor)
            contador += 1

#Ya hemos recorrido el array entero, así que la variable contador ya es igual
#al número de elementos del array que han cumplido la condición.
print(f"\nEl número de elementos que cumplen la condición es: {contador}")
```

### **Ejercicio 2**

Genera los siguientes arrays utilizando **numpy**:

- **a)** Un array 1D de **15 elementos**, donde los valores en posiciones **múltiplos de 3** sean `0` y el resto `5`.  
- **b)** Un array 1D con los **números pares del 10 al 28** (inclusive), ordenados **ascendentemente**.  
- **c)** Un array 2D de dimensiones **3x4**, con valores aleatorios **enteros** entre 50 y 100.  
- **d)** Un array 2D de dimensiones **2x6**, con los valores del 1 al 12 en **orden descendente**, utilizando la función `reshape()`.

**(0.5 puntos)**  
$\color{green}{\text{NM}}$

```python
# importamos libs
import numpy as np
# Respuesta A
#Creamos un array de 15 elementos en una sóla fila donde todos los elementos tengan
#valor 5.
array_a = np.full(15, 5)
#Convertirmos los valores en posiciones múltiplos de 3 en 0 utilizando un for.
#Se utiliza i+1 porque se ha supuesto que cuando el enunciado habla de posiciones
#se refiere a las posiciones reales y no a la numeración de python que empieza por
#0. Si no fuera así, basta con eliminar (i+1) y dejar sólo i, o que haría al primer
#elemento (índice 0) también 0, ya que 0 % 3 == 0.
for i in range(array_a.size):
  if (i+1) % 3 == 0:
        array_a[i] = 0
#Imprimimos el array resultante
print(array_a)
```

```python
# ---------------------------------------------------------------------
# b) Array 1D con números pares del 10 al 28 inclusive (ascendente)
# ---------------------------------------------------------------------

#Utilizaremos np.arange() para generar la secuencia. Indicamos como punto de
#del array 10, luego le damos como punto final el 29 (ya que el final está
#excluído) y por último paso 2 para que sólo genere los pares.

array_b = np.arange(10, 29, 2)
print(array_b)
```

```python
# ---------------------------------------------------------------------
# c) Array 2D (3x4) con valores enteros aleatorios entre 50 y 100
# ---------------------------------------------------------------------

#Creamos un array con valores enteros aleatorios entre 50 y 100, utilizamos 101
#ya que el límite vuelve a no estar incluído.

#El último elemento de randint indica las dimensiones del array.
array_c = np.random.randint(50, 101, (3, 4))
print(array_c)
```

```python
# ---------------------------------------------------------------------
# d) Array 2D (2x6) con valores del 1 al 12 en orden descendente, usando reshape
# ---------------------------------------------------------------------
#Creamos el array, está vez con paso -1 para que el orden sea descendente.
array_d = np.arange(12, 0, -1)
print(array_d)

#Como vemos, hemos creado un array unidimensional. Ahora usamos reshape para
#darle las dimensiones requeridas.
array_d = array_d.reshape(2, 6)
print(f"\n Array 2x6:\n {array_d}")
```

### **Ejercicio 3**

Crea un array bidimensional de ceros de dimensiones **(12, 12)**, y modifica los valores necesarios para representar los siguientes patrones:

- **a)** Un **marco** (bordes del array) formado por 1s.   **(0.5 puntos)**  
- **b)** Una **diagonal secundaria** (de la esquina superior derecha a la inferior izquierda) formada por 1s.   **(0.5 puntos)**  
- **c)** Una **cuadrícula** con líneas horizontales y verticales cada 3 posiciones.    **(0.5 puntos)**  
- **d)** Un **patrón de triángulo** con 1s en la parte **inferior izquierda** de la matriz.    **(0.5 puntos)**  

**Nota:** Utiliza una función de **`matplotlib`** para visualizar cada patrón como una imagen 2D (por ejemplo, con escala de grises).

A modo de ejemplo del resultado esperado, se proporcionan las siguientes figuras:

FiguraEj3.png

**(Total: 2 puntos)**  
$\color{orange}{\text{EG}}$

```python
# importo librerías
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Respuesta A
#Creamos el array de 0s.
cuadrado_a = np.zeros((12, 12))

#Relenamos bordes asignando el valor 1 a la primera y última fila y a la
#primera y última columna usando slicing.
cuadrado_a[0, :] = 1
cuadrado_a[-1, :] = 1
cuadrado_a[:, 0] = 1
cuadrado_a[:, -1] = 1

#Utilizamos la función .imshow() para representar el array cuadrado_a como imagen 2D.
#El argumento cmap = gray nos dará la escala de grises que sugiere el enunciado.
#Esto significa que los valores 0 se muestran en negro y los 1 en blanco.
plt.imshow(cuadrado_a, cmap='gray')
#Utilizamos.axis() para que no aparezcan los ejes de la representación y quede
#más limpio.
plt.axis('off')
plt.show()
```

#http://matplotlib.org/examples/color/colormaps_reference.html

```python
# Respuesta B
cuadrado_b = np.zeros((12, 12))

#Con un for asignamos a cada elemento de la diagonal secundaria el valor 1
for i in range(cuadrado_b.shape[0]):
  #Al número de columnas le restamos 1 e i
  #.shape[1] devuelve el número de columnas total, al que hay que restar 1 para
  #seguir la numeración Python, se resta i ya que es la diagonal secundaria y así
  #avanzamos de última columna a primera.
    cuadrado_b[i, cuadrado_b.shape[1]-1-i] = 1

#Representamos el cuadrado.
plt.imshow(cuadrado_b, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Respuesta C
cuadrado_c = np.zeros((12, 12))

#Utilizaremos slicing para asignar a los elementos de las filas y columnas cada tres
#posiciones el valor 1. ::3 es de inicio a fin con paso 3.
cuadrado_c[::3, :] = 1
cuadrado_c[:, ::3] = 1


plt.imshow(cuadrado_c, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Respuesta D
cuadrado_d = np.zeros((12, 12))

#Utilizamos un for para crear el patrón del triángulo inferior izquierdo.
#En cada fila i, asignamos el valor 1 a las columnas desde 0 hasta i usando slicing
#[:i+1] va desde el inicio hasta i (ya que el final está exlcuído, de ahí
#el i+1).

for i in range(cuadrado_d.shape[0]):
    cuadrado_d[i, :i+1] = 1

plt.imshow(cuadrado_d, cmap='gray')
plt.axis('off')
plt.show()
```

### Ejercicio 4

Imagina que tienes tres cuerpos en un plano bidimensional.  Las fuerzas que actuan sobre cada cuerpo se pueden representar en una matriz.  Cada fila de la matriz representa un cuerpo y cada columna representa una fuerza en una dirección concreta ( per ejemplo, fuerza en la dirección X y fuerza en la dirección Y) en Newtons.

Las fuerzas que actuan sobre los cuerpos son las seguientes:
- Cuerpo A: Fuerza X = 10 y Fuerza Y = 5.
- Cuerpo B: Fuerza X = 4 y Fuerza Y = 12.
- Cuerpo C: Fuerza X = -6 y Fuerza Y = -8.

Responde a las seguientes preguntas:
- a) Crea la matriz de fuerzas y muéstrala por pantalla. **(0.25 puntos)**
- b) Calcula la fuerza resultante sobre cada cuerpo. La fuerza resultante se puede calcular como la suma vectorial de las fuerzas.  Para hacerlo, utiliza una función de `numpy`. **(0.5 puntos)**
- c) Calcula la aceleración de cada cuerpo mediante la segunda ley de Newton, `F=m*a`, considerando que la masa de cada cuerpo es de 5 kg. **(0.25 puntos)**

**Nota**.  Los resultados deben presentarse con sólo dos decimales.

**(Total 1 punto)** $\color{Orange}{\text{EG}}$

```python
# Apartado a) creamos la matriz de fuerzas

#Ya tenemos las librerías importadas así que lo obviamos aquí.
#Creamos la matriz dándole forma.
matriz = np.array([[10,  5], [ 4, 12], [-6, -8]])
print(f"Matriz de fuerzas:\n {matriz}")
```

```python
# Apartado b) Calculamos la fuerza resultante
#Utilizamos linalg.norm, es decir, el sub-módulo linalg que nos permite calcular
#la norma euclidiana (es decir, la longitud/distancia) del vector de fuerzas.
#Axis=1 para que calcule la norma de cada fila, si usáramos axis=0, calcularía la
#norma de cada columna y si no especificaramos axis calcularía la norma de toda
#la matriz.
fuerza = np.linalg.norm(matriz, axis=1)
print(fuerza)

#Vemos cómo hemos obtenido un array con las fuerzas resultantes sobre cada
#cuerpo, ahora imprimimos los resultados por separado y con dos decimales.
print(f"Cuerpo A: {fuerza[0]:.2f} N")
print(f"Cuerpo B: {fuerza[1]:.2f} N")
print(f"Cuerpo C: {fuerza[2]:.2f} N")
```

#https://numpy.org/doc/stable/reference/routines.linalg.html

```python
# Apartado c)
masa = 5
#Como ya tenemos el array de fuerzas, se puede divir por escalar (masa) y se
#aplica la operación a cada elemento del array.
aceleracion = fuerza / masa
print(aceleracion)
#Ahora sólo tenemos que imprimirlos por separado(para el ² usamos alt + 0178)
print(f"Cuerpo A: {aceleracion[0]:.2f} m/s²")
print(f"Cuerpo B: {aceleracion[1]:.2f} m/s²")
print(f"Cuerpo C: {aceleracion[2]:.2f} m/s²")
```

```python
# Respuesta A
#Importamos las nuevas libs, utilizaremos scipy para generar los valores en vez
#de numpy
import seaborn as sns
from scipy import stats


#Utilizaremos stats.norm.rvs para generar los valores. loc hace referencia a la
#media, scale la desviación estándar y size al número de valores a generar.
datos_normal = stats.norm.rvs(loc=50, scale=10, size=100)
```

```python
# Respuesta B: Test de normalidad
shapiro_test = stats.shapiro(datos_normal)
p_val = shapiro_test.pvalue
print(f"p-value = {p_val:.4f}")

# El test de Shapiro-Wilk compara los datos que le suministramos
#con una distribución normal, si el p-value resulta mayor que 0.05, los datos
#siguen una distribución normal (realmente, no podemos rechazar la hipótesis nula H0
#de que los datos siguen una distribución normal)

if shapiro_test.pvalue > 0.05:
    print(f"\np-value > 0.05, los datos siguen una distribución normal.")

else:
    print(f"\np-value <= 0.05, no podemos asumir que los datos siguen una distribución normal.")

```

```python
# Respuesta C:  Histograma de la distribución
#Creamos el histograma, utilizaremos matplotlib. Asignamos 10 intervalos como
#en el histograma anterior y damos algo de formato al histograma para que sea más visual.
plt.hist(datos_normal, bins=10, facecolor='steelblue', alpha=0.7, edgecolor='black')
plt.title('Distribución de datos generados')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Respuesta D: (Opcional) Visualización con seaborn
#Como ya importamos seaborn al principio del ejercicio, vamos directos a utilizar
#kdeplot, creando una curva (la cual rellenaremos con fill=True) para estimar visualmente
#si los datos generados siguen una distribución normal.
sns.kdeplot(datos_normal, fill=True, color='steelblue', alpha=0.6, label='Datos generados')
plt.title('Curva de densidad de probabilidad')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```
